```python
import warnings
warnings.filterwarnings('ignore')
import yfinance as yf
import numpy as np
import pandas as pd
#define the ticker symbol
tickerSymbol = 'UBER'

#get data on this ticker
tickerData = yf.Ticker(tickerSymbol)

#get the historical prices for this ticker
dfu = tickerData.history(period='1d', start='2019-5-10', end='2022-10-18')

#see your data
dfu
```

```python
#define the ticker symbol
tickerSymbol = 'LYFT'

#get data on this ticker
tickerData = yf.Ticker(tickerSymbol)

#get the historical prices for this ticker
dfl = tickerData.history(period='1d', start='2019-5-10', end='2022-10-18')

#see your data
dfl
```

```python
import matplotlib.pyplot as plt   # Import matplotlib
# This line is necessary for the plot to appear in a Jupyter notebook
%matplotlib inline
# Control the default size of figures in this Jupyter notebook
%pylab inline

pylab.rcParams['figure.figsize'] = (15, 9)   # Change the size of plots

dfu["Close"].plot(label='Uber Technologies', color = '#000000')
dfl["Close"].plot(label='Lyft', color='#FF00BF')

# adding title to the plot
plt.title('Close Price Plot for Uber Technologies & Lyft')
# adding Label to the x-axis
plt.xlabel('Date')
# adding legend to the curve
plt.legend()
```

```python
close_u = dfu["Close"]
close_l = dfl["Close"]

simple_return_u = close_u.pct_change()
simple_return_l = close_l.pct_change()

log_return_u = np.log(1+simple_return_u)
log_return_l = np.log(1+simple_return_l)
```

```python
pylab.rcParams['figure.figsize'] = (15, 9)   # Change the size of plots

simple_return_u.plot(label='Uber Technologies', color = '#000000')
simple_return_l.plot(label='Lyft', color='#FF00BF')

# adding title to the plot
plt.title('Simple Return for Uber Technologies & Lyft')
# adding Label to the x-axis
plt.xlabel('Date')
# adding legend to the curve
plt.legend()
```

```python
pylab.rcParams['figure.figsize'] = (15, 9)   # Change the size of plots

log_return_u.plot(label='Uber Technologies', color = '#000000')
log_return_l.plot(label='Lyft', color='#FF00BF')

# adding title to the plot
plt.title('Log Return for Uber Technologies & Lyft')
# adding Label to the x-axis
plt.xlabel('Date')
# adding legend to the curve
plt.legend()
```

```python
import seaborn as sns

simple_return1_u = simple_return_u[np.logical_not(np.isnan(simple_return_u))]
simple_return1_l = simple_return_l[np.logical_not(np.isnan(simple_return_l))]

# histogram
sns.distplot(simple_return1_u, kde=False, norm_hist=True,label='Simple returns Uber Technologies', color = '#000000') 
sns.distplot(simple_return1_l, kde=False, norm_hist=True, label='Simple returns Lyft', color='#FF00BF')  

legend(loc='upper left');

plt.show()
```

```python
print('---------- Descriptive statistics simple returns Uber Technologies ----------')
print('Number of observations:', simple_return1_u.shape[0])
print(f'Mean: {simple_return1_u.mean():.4f}')
print(f'Median: {simple_return1_u.median():.4f}')
print(f'Min: {simple_return1_u.min():.4f}')
print(f'Max: {simple_return1_u.max():.4f}')
print(f'Standard Deviation: {simple_return1_u.std():.4f}')
print(f'Skewness: {simple_return1_u.skew():.4f}')
print(f'Kurtosis: {simple_return1_u.kurtosis():.4f}') 

print('')
print('---------- Descriptive statistics simple returns Lyft ----------')
print('Number of observations:', simple_return1_l.shape[0])
print(f'Mean: {simple_return1_l.mean():.4f}')
print(f'Median: {simple_return1_l.median():.4f}')
print(f'Min: {simple_return1_l.min():.4f}')
print(f'Max: {simple_return1_l.max():.4f}')
print(f'Standard Deviation: {simple_return1_l.std():.4f}')
print(f'Skewness: {simple_return1_l.skew():.4f}')
print(f'Kurtosis: {simple_return1_l.kurtosis():.4f}') 
```

```python
log_return1_u = log_return_u[np.logical_not(np.isnan(log_return_u))]
log_return1_l = log_return_l[np.logical_not(np.isnan(log_return_l))]

# histogram
sns.distplot(log_return1_u, kde=False, norm_hist=True,label='Log returns Uber Technologies', color = '#000000') 
sns.distplot(log_return1_l, kde=False, norm_hist=True, label='Log returns Lyft', color='#FF00BF')  

legend(loc='upper left');

plt.show()
```

```python
print('---------- Descriptive statistics log returns Uber Technologies ----------')
print('Number of observations:', log_return1_u.shape[0])
print(f'Mean: {log_return1_u.mean():.4f}')
print(f'Median: {log_return1_u.median():.4f}')
print(f'Min: {log_return1_u.min():.4f}')
print(f'Max: {log_return1_u.max():.4f}')
print(f'Standard Deviation: {log_return1_u.std():.4f}')
print(f'Skewness: {log_return1_u.skew():.4f}')
print(f'Kurtosis: {log_return1_u.kurtosis():.4f}') 

print('')
print('---------- Descriptive statistics log returns Lyft ----------')
print('Number of observations:', log_return1_l.shape[0])
print(f'Mean: {log_return1_l.mean():.4f}')
print(f'Median: {log_return1_l.median():.4f}')
print(f'Min: {log_return1_l.min():.4f}')
print(f'Max: {log_return1_l.max():.4f}')
print(f'Standard Deviation: {log_return1_l.std():.4f}')
print(f'Skewness: {log_return1_l.skew():.4f}')
print(f'Kurtosis: {log_return1_l.kurtosis():.4f}') 
```
